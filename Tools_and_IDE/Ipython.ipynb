# Ipython（QTconsole）：交互式Python控制台程序

## Ipython简介

IPython（Interactive Python）由Fernando Perez于2001年创建，旨在为Python提供更强大的交互式外壳。该工具基于Mathematica的笔记本界面进行演进，其设计目标是提升交互式计算体验。

多年来，IPython已成为Jupyter笔记本及更广泛的Jupyter项目生态系统的基石，彻底改变了科学家、数据分析师和教育工作者使用代码的方式。

Ipython是Python在控制台界面之上的一个全新环境，既是编辑器的重大升级，也是很多软件的基础。与控制台Python不同的是，其命令输入和输出之前的符号不再是`>>>`，而是`In[]`作为标识。



## IPython 的主要功能

**交互式 Shell**：IPython 提供了比默认 Python shell 更加友好的交互环境，支持变量自动补全和自动缩进，使得编写和调试代码更加高效。

**内省功能**：用户可以通过在对象后加上问号（?）来获取对象的详细信息，使用双问号（??）可以查看对象的源代码。这种内省能力使得用户能够快速了解函数和类的用法。

**魔法命令**：IPython 提供了多种魔法命令（magic commands），如 %run 用于运行 Python 脚本，! 用于执行系统命令等。这些命令可以简化常见操作，提高工作效率。

**支持 Shell 命令**：用户可以直接在 IPython 中执行 Bash shell 命令，方便进行系统级操作。

**历史记录**：IPython 自动保存输入的历史记录，用户可以方便地查看和重用之前的命令。


## QTconsole:一种基于QT的轻量级Python工作台

Ipython是基于命令行的工具，但在图形现实等方面，缺乏必要支撑，于是一款基于Qt的丰富控制台，支持Jupyter内核操作，具备多媒体输出、会话导出等功能的工具在其基础上被创造出来，就是QTconsole。

Qtconsole是一款轻量级应用程序，其界面设计与终端高度相似，但具备图形用户界面（GUI）独有的多项增强功能，包括内嵌图表、支持语法高亮的多行文本编辑、图形化调用提示等。



QTconsole项目由Spyder开发团队负责维护，并隶属于其组织架构。
